```python
import joblib

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Red neuronal secuencial
from tensorflow.keras.models import Sequential
#Capa completamente conectada
from tensorflow.keras.layers import Dense, Normalization
#Optimizador
from tensorflow.keras.optimizers import Adam
import tensorflow as tf


from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('data.csv', encoding='unicode_escape')
data.columns = ['Estanque', 'Hectareas', 'tC_am', 'tC_pm','O2_am', 'O2_pm', 'Biomasa', 'Dias_cultivo', 'Dia_anio', 'Total']
df = pd.DataFrame(data)

df = df.loc[:,['Estanque', 'Hectareas', 'tC_am', 'O2_am', 'Biomasa', 'Dias_cultivo', 'Dia_anio', 'Total']]

df = df.dropna()
df
```

```python
scaler = MinMaxScaler().fit(df)
scaler
```

```python
data_norm = scaler.transform(df)

data_norm_df = pd.DataFrame(data_norm, columns=['Estanque', 'Hectareas', 'tC_am', 'O2_am', 'Biomasa', 'Dias_cultivo', 'Dia_anio', 'Total'])

data_norm_df
```

```python
modelin = joblib.load('intento1.pkl')
```

```python
def nuevosValores(params):

  parametros = pd.DataFrame({'Estanque': [params[0]], 'Hectareas': [params[1]], 'tC_am': [params[2]], 'O2_am': [params[3]], 'Biomasa': [params[4]],'Dias_cultivo': [params[5]],'Dia_anio': [params[6]], 'Total': [None]})
  nuevoDf_norm = scaler.transform(parametros)

  valoresNorm = nuevoDf_norm[0][0:7]

  return valoresNorm
```

```python
def prediccionIt(prd):

  prediccion=pd.DataFrame({'Estanque': [None], 'Hectareas': [None], 'tC_am': [None], 'O2_am': [None], 'Biomasa': [None],'Dias_cultivo': [None],'Dia_anio': [None], 'Total': [prd]})
  nuevo_it = scaler.inverse_transform(np.array(prediccion))

  prediccionIt = nuevo_it[0][-1]

  return prediccionIt
```

```python
def prediccionDef (params):
  i = nuevosValores(params)
  i = i.tolist()
  prediccion = modelin.predict([i])
  valorPredicho = prediccionIt(prediccion)

  return valorPredicho
```
